# 9. Lists and List Operations

A list stores multiple values in one variable.

Lists are used often in Python because data usually comes in groups.

````python
stocks = ["AAPL", "MSFT", "GOOG","AMZN"]

print(stocks)
print(stocks[0])
print(stocks[-1])
````

## Common List Operations

| Operation | Example |
|---|---|
| add item | `append()` |
| insert item | `insert()` |
| remove item by value | `remove()` |
| remove item by position | `pop()` |
| sort list | `sort()` |
| reverse list | `reverse()` |
| count items | `len()` |
| sum numbers | `sum()` |
| largest value | `max()` |
| smallest value | `min()` |

````python
prices = [120, 95, 310, 450, 80]

print("Original:", prices)

prices.append(200)
print("After append:", prices)

prices.insert(1, 150)
print("After insert:", prices)

prices.remove(95)
print("After remove:", prices)

last_price = prices.pop()
print("Removed last item:", last_price)
print("After pop:", prices)

prices.sort()
print("Sorted:", prices)

prices.reverse()
print("Reversed:", prices)

print("Length:", len(prices))
print("Sum:", sum(prices))
print("Max:", max(prices))
print("Min:", min(prices))
print("Average:", sum(prices) / len(prices))
````

## List Slicing

# 17. Classes and Objects

Classes are used to create custom types of objects.

A class is like a blueprint.

An object is something created from that blueprint.

Example:
- Class: Student
- Objects: Maya, Aarav, Riya

## Basic Class Example

````python
class Student:
    def __init__(self, name, age, score):
        self.name = name
        self.age = age
        self.score = score

student1 = Student("Maya", 16, 92)

print(student1.name)
print(student1.age)
print(student1.score)
````

````python
class Car:
  def __init__(self,brandName, colour, size):
    self.name = brandName
    self.colour1 = colour
    self.size = size
car1 = Car("Mercedes","Black",8)
print(car1.name)
print(car1.colour1)
print(car1.size)
````

## Adding Methods to a Class

A method is a function inside a class.

````python
class Student:
    def __init__(self, name, age, score):
        self.name = name
        self.age = age
        self.score = score

    def display_info(self):
        print(self.name, "is", self.age, "years old and scored", self.score)

    def passed(self):
        return self.score >= 50

student1 = Student("Maya", 16, 92)
student2 = Student("Aarav", 15, 45)

student1.display_info()
student2.display_info()

print(student1.passed())
print(student2.passed())
````
